### LIST NILAI
Diberikan sekumpulan Nilai Random antara 0 sampai dengan 100 sebanyak 250 nilai. Buatlah program untuk melakukan hal berikut:

1. Tampilkan Angka Ganjil dari ke 50 nilai tersebut
2. Tampilkan berapa banyak yang mendapatkan nilai lebih dari 70
3. Tampilkan standar deviasi dari sekumpulan nilai tersebut
4. Tampilkan nilai maksimum dan minimum
5. Tampilkan modus dari sekumpulan nilai tersebut

```python
import random

# GENERATE RANDOM
nilaiTugas = []
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))

print(nilaiTugas)

# TAMPILKAN:
# ganjil
# nilai > 70
# standar deviasi
# maksimum
# minimum
# modus
```

```python
#menampilkan nilai ganjil
for i in nilaiTugas:
    if i % 2:
        print(i)
```

```python
#menampilkan nilai lebih dari 70
for i in nilaiTugas:
    if i > 70:
        print(i)
```

```python
#menghitung standar deviasi
jumlah = 0
for i in range(len(nilaiTugas)):
    jumlah += nilaiTugas[i]    
print("Jumlah = ",jumlah)

rata=jumlah/len(nilaiTugas)
print("Rata rata = ",rata)

sigma = 0
for i in range(len(nilaiTugas)):
    hitung =(nilaiTugas[i]-rata)**2
    sigma += hitung

print("Sigma = ",sigma)

pembagianN = sigma/len(nilaiTugas)
standardeviasi = pembagianN ** 0.5
print("Standar Deviasi = ",standardeviasi)
```

```python
#nilai max & min

nilaiMin = 100
nilaiMax = 0
for nilai in nilaiTugas:
    if nilai < nilaiMin:
        nilaiMin = nilai
    if nilai > nilaiMax:
        nilaiMax = nilai
        
        
print("nilai Maximum =", nilaiMin)
print("nilai Maximum =", nilaiMax)


```

```python
#nilai Modus
nilaiDict = dict()

for nilai in nilaiTugas:
    nilaiDict[nilai] = nilaiDict.get(nilai,0) + 1
    
ListNilai = list()
for nilai, count in nilaiDict.items():
    ListNilai.append((count, nilai))
    
ListNilai.sort(reverse = True)
print("Nilai Modus: ", ListNilai[0][1], " jumlahnya: ", ListNilai[0][0])    
```
